```python
class Base_Salary:
  def __init__(self, base_salary, bonus_rate=0.1, symbols="$"):
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbols = symbols
    self.total_salary = self.base_salary * (1 + self.bonus_rate)
    self.bonus = (self.total_salary - base_salary)/base_salary

  def __repr__(self):
    return f"{self.symbols}{self.base_salary:,.0f}"

  def calculate_salary(self):
    return f'{self.symbols}{self.total_salary:,.0f}'

  def calculate_bonus(self):
    return f'{self.symbols}{self.bonus:,.0f}'
```

```python
salary = Base_Salary(10000)
salary.base_salary
salary.symbols
```

```python
salary
```

```python
salary.calculate_salary()
```

```python
salary.calculate_bonus()
```
